```
import numpy as np

def activation_function(value):
  if value >= 0:
    return 1
  else:
    return 0

# columns must be = to the number of values in weights
def predict(row_data, weights, bias):
  return activation_function(np.sum(np.add(np.dot(row_data, weights.T), bias)))
```

```
# Create dataset
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv')
df
```

```
weights = np.asarray([.5,.5,.5,.5,.5,.5,.5,.5])
bias = np.asarray([.5])


X_vars = np.asarray(df.iloc[:,:-1])

accuracy = []

bias = np.asarray([.5])

def train(weights = weights, bias = bias, epochs = 50):
  for e in range(epochs):
    print('[',e,'/',epochs,']')
    row_number = 0
    number_correct = 0
    for row in X_vars:
      #print(predict(row, weights, bias), np.asarray(df.iloc[row_number:row_number+1,-1:]))
      error = predict(row, weights, bias) - df.iloc[row_number:row_number+1,-1:]
      if np.asscalar(np.asarray(error)) == 0:
        number_correct += 1
      bias -= 0.001 * np.asscalar(np.asarray(error))
      weights -= 0.001 * np.asscalar(np.asarray(error)) * row
      row_number += 1
    # print(weights)
    accuracy.append(number_correct / row_number)

train()
```
